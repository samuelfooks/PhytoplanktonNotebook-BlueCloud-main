# Acquire BlueCloud Data products 

https://www.blue-cloud.org/demonstrators/zoo-and-phytoplankton-eov-products

Zoo-Phytoplankton_EOV >Phytoplankton_EOV >Chla_Product >Outputs >20210222

Inside are all the outputs used in the BlueCloud 

Save in folder called 'netcdfs'

```python
import xarray as xr
import zarr
import os

# Directory containing the NetCDF files
directory = 'netcdfs'

# List all NetCDF files in the directory
files = [file for file in os.listdir(directory) if file.endswith('.nc')]

# Initialize a Zarr store
store = zarr.DirectoryStore('phytoplankton/soca_chla_01_12.zarr')

# Process the NetCDF files one by one
for i, file in enumerate(files):
    # Open the NetCDF file as a dataset
    dataset = xr.open_dataset(os.path.join(directory, file))
    
    # If this is the first file, write it to the Zarr store
    # Otherwise, append the dataset to the Zarr store
    if i == 0:
        dataset.to_zarr(store, mode='w')
    else:
        dataset.to_zarr(store, mode='a', append_dim='time')
```

Check that it worked

```python
combinedds = xr.open_dataset('phytoplankton/soca_chla_01_12.zarr', engine='zarr')

print(combinedds['time'].min())

print(combinedds['time'].max())
```
